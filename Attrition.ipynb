Data Visualization

```python
#Distribution of attrition
plt.figure(figsize=(6, 4))
ax= sns.countplot(x='Attrition', data=df, palette=['skyblue', 'salmon'])
for i in ax.containers:
    ax.bar_label(i)
plt.title('Attrition Distribution')
plt.xlabel('Attrition')
plt.ylabel('Count')
plt.show()
```

1.The vast majority of employees have not left the company, with a count of 1233.
2.A smaller number of employees, 237, have left the company.
3.From this visual we can make out that attrition is relatively low.

```python
#Distribution of age
plt.figure(figsize=(6, 4))
sns.histplot(df['Age'], bins=20, kde=True)
plt.title('Age Distribution')
plt.xlabel('Age')
plt.ylabel('Frequency')
plt.show()
```

1.The peak around 30-35 could suggest that this is the age group where most employees are hired or retained. 
2.The right skew might indicate that there are a higher number of older employees compared to younger ones. 
This could be due to factors like career progression, longer tenure, or industry-specific trends.

```python
#Age Vs attrition
plt.figure(figsize=(4, 6))
sns.boxplot(y='Age', x='Attrition', data=df, palette='Set2')
plt.title('Age Vs attrition')
plt.xlabel('Attrition')
plt.ylabel('Age')
plt.show()
```

1.The median age for employees who have left is lower(around 30) than those who have stayed(around 37-38). 
2. This box plot suggests that younger employees are more likely to leave the company compared to older employees.

```python
#Monthly income by attrition
plt.figure(figsize=(4, 6))
sns.boxplot(x='Attrition', y='MonthlyIncome', data=df, palette='Set2')
plt.title('Monthly Income by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Monthly Income')
plt.show()
```

1.The employees who have not left the company has a wider spread of incomes, with the box extending from 
around 5000 to 9000. Whereas the employees who have left the company have a narrower spread, with the box spanning roughly from 4000 to 6000.
2.This suggests that employees with lower monthly incomes are more likely to leave the company compared to those with higher incomes

```python
#Daily wages and Attrition
plt.figure(figsize=(4, 6))
sns.boxplot(x='Attrition', y='DailyRate', data=df, palette='Set2')
plt.title('Daily Wages by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Daily Rate')
plt.show()
```

1. The employees who have not left the company has a wider spread of wages, with the box extending from around 600 to 1000.
Whereas the employees who have left the company  has a narrower spread, with the box spanning roughly from 400 to 800.
2. This suggests that employees with lower daily wages are more likely to leave the company compared to those with higher wages.

```python
#Hourly Rate and Attrition
plt.figure(figsize=(4, 6))
sns.boxplot(x='Attrition', y='HourlyRate', data=df, palette='Set2')
plt.title('Hourly Rate by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Hourly Rate')
plt.show()
```

```python
#Total Working Years vs Monthly Income
plt.figure(figsize=(8, 6))
sns.scatterplot(x='TotalWorkingYears', y='MonthlyIncome', hue='Attrition', data=df, palette='coolwarm')
plt.title('Total Working Years vs Monthly Income')
plt.xlabel('Total Working Years')
plt.ylabel('Monthly Income')
plt.show()
```

1.The "No" group (employees who stayed) has a broader range of monthly incomes across all working years. 
There are employees with very high incomes even with fewer working years, and vice versa.
2.The "Yes" group (employees who left) appears to have a more concentrated range of monthly incomes, 
with a majority of employees having lower to moderate incomes across all working years.
3. This scatter plot suggests that while both groups tend to see an increase in monthly income 
with more working years, the employees who have left the company generally have lower monthly incomes compared to those who have stayed,
especially for similar levels of experience.

```python
#Year at Company by Attrition
plt.figure(figsize=(4, 6))
ax=sns.boxplot(x='Attrition', y='YearsAtCompany', data=df, palette='Set2')
plt.title('Years at Company by Attrition')
plt.xlabel('Attrition')
plt.ylabel('Years at Company')
plt.show()
```

```python
#Distance from residence Vs Attrition
plt.figure(figsize=(4, 6))
ax=sns.barplot(x='Attrition', y='DistanceFromHome', data=df, palette='coolwarm',errorbar=None)
for i in ax.containers:
    ax.bar_label(i)
plt.title('Distance From Residence')
plt.xlabel('Attrition')
plt.ylabel('Distance From Residence by Attrition')
plt.show()
```

```python
#Attritionvs Years Since Last Promotion
plt.figure(figsize=(8, 6))
sns.lineplot(x='YearsSinceLastPromotion', y='Attrition', data=df, markers=True)
plt.title('Attrition vs Years Since Last Promotion')
plt.xlabel('Years Since Last Promotion')
plt.ylabel('Attrition Rate')
plt.show()
```

1. While the overall trend is upward, there are significant fluctuations in the attrition rate at different points.

Level 4: The count of employees who have left decreases again compared to Level 3.
This bar chart suggests that employees with lower job involvement (Levels 1 and 2)

This implies that employees who are less engaged and committed to their work are more likely to seek opportunities elsewhere.

```python
fig,ax = plt.subplots(2,3, figsize=(14,14))               # 'ax' has references to all the four axes
plt.suptitle("Comparision of various factors vs Gender", fontsize=20)
sns.barplot(x='Gender',y='DistanceFromHome',hue = 'Attrition',data=df, ax = ax[0,0],errorbar=None, palette='coolwarm'); 
sns.barplot(x='Gender',y='YearsAtCompany',hue = 'Attrition', data=df,ax = ax[0,1],errorbar=None, palette='coolwarm'); 
sns.barplot(x='Gender',y='TotalWorkingYears',hue = 'Attrition',data=df, ax = ax[0,2],errorbar=None, palette='coolwarm'); 
sns.barplot(x='Gender',y='YearsInCurrentRole',hue = 'Attrition', data=df,ax = ax[1,0],errorbar=None, palette='coolwarm'); 
sns.barplot(x='Gender',y='YearsSinceLastPromotion',hue = 'Attrition',data=df, ax = ax[1,1],errorbar=None, palette='coolwarm'); 
sns.barplot(x='Gender',y='NumCompaniesWorked',hue = 'Attrition', data=df,ax = ax[1,2],errorbar=None, palette='coolwarm'); 
plt.show()
```

Distance From Home: Females tend to live slightly farther from work than males, but their attrition rate remains lower regardless of distance.
Years at Company: Males generally have more years at the company, and their attrition rates decrease as tenure increases.
Total Working Years: Males have a higher total working experience on average, with lower attrition as working years increase.
Years in Current Role: Males stay longer in their current roles compared to females, and attrition decreases with increased years in the role.
Years Since Last Promotion: Males show slightly higher years since the last promotion, but attrition is higher for 
males with longer periods without promotion.
Number of Companies Worked: Males have worked at more companies than females, and attrition is higher 
for males with a greater number of previous employers.
While males tend to have longer tenures and more extensive work histories, females consistently exhibit lower attrition rates, 
indicating greater job stability across these factors.
